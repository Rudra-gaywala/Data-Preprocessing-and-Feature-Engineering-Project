# Part A: Theory - Data Analysis & Machine Learning Fundamentals


## 1. What is Data Analysis?
Data analysis is the process of collecting, cleaning, transforming, and interpreting data to extract meaningful insights and support decision-making.

---

## 2. Steps in a Data Science Project
1. Problem Definition  
2. Data Collection  
3. Data Cleaning  
4. Exploratory Data Analysis (EDA)  
5. Feature Engineering  
6. Model Building  
7. Evaluation  
8. Deployment  

---

## 3. Machine Learning Problem Statement
The goal is to predict whether a customer will churn based on:
- Demographics  
- Purchase behavior  
- Income  
- Engagement patterns  

This is a **classification problem**.

---

## 4. What are Tensors?
Tensors are multi-dimensional arrays used in machine learning and deep learning.

### Types of Tensors:
- Scalar (0D): Single value  
- Vector (1D): List of values  
- Matrix (2D): Table of values  
- Higher-order tensors (3D+): Used in deep learning  

---

## 5. Importance of Data Preprocessing
Data preprocessing ensures:
- Clean data  
- No missing values  
- Proper formatting  
- Better model performance  

---

## 6. Types of Data
- Structured Data (tables, CSV)  
- Semi-structured Data (JSON, XML)  
- Unstructured Data (text, images, videos)  

---

## 7. Exploratory Data Analysis (EDA)
EDA is used to:
- Understand data distribution  
- Detect outliers  
- Find relationships between variables  

Types:
- Univariate  
- Bivariate  
- Multivariate  

---

## 8. Feature Engineering
Feature engineering involves:
- Creating new features  
- Transforming existing features  
- Encoding categorical variables  

---

## 9. Conclusion
Data analysis and preprocessing are essential steps in any machine learning pipeline. Proper handling of data leads to better insights and improved model accuracy.
